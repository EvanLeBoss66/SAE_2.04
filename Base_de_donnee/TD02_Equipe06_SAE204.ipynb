```python
import pandas as pd
import matplotlib.pyplot as plt

df_tenrac = pd.read_csv("tenrac.csv", sep=";")
valeurs = df_tenrac['ID_GENRE'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(valeurs, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.legend(["Femme", "Homme", "Autre"], title="Genres")
plt.title("Répartition Homme / Femme / Autre")
plt.show()
```

# Ce graphique montre la fréquentation selon le genre. Il indique que la clientèle est assez équilibrée, sans grande différence entre les sexes. Cela ne nous aide pas vraiment à changer la carte, mais montre que la plupart sont satisfaits, peu importe leur genre. Le choix du graphique en camembert est idéal ici, car il affiche clairement les proportions et est facile à lire, avec des pourcentages visibles.

```python
import pandas as pd
import matplotlib.pyplot as plt

df_tenrac = pd.read_csv("tenrac.csv", sep=";")
df_lien = pd.read_csv("allergique_a_.csv", sep=";")
df_allergies = pd.read_csv("allergies.csv", sep=";")

df_allergies = df_allergies.rename(columns={'NOM': 'NOM_ALLERGIE'})
df_fusion = df_tenrac.merge(df_lien, on="ID_MEMBRE").merge(df_allergies, on="ID_ALLERGENE")

top = df_fusion.groupby(['ID_GENRE', 'NOM_ALLERGIE']).size().reset_index(name='Total')
top = top.loc[top.groupby('ID_GENRE')['Total'].idxmax()]

plt.figure(figsize=(10, 6))
graph = plt.bar(top['NOM_ALLERGIE'], top['Total'], color=['#ff9999', '#66b3ff', '#99ff99'])

plt.legend(graph, ["Femme", "Homme", "Autre"], title="Genres")
plt.title("Allergie la plus fréquente par sexe")
plt.show()
```

# Ce graphique met en évidence l'allergie la plus répandue pour chaque sexe. Il aide à comprendre les contraintes de santé des membres et à adapter les menus pour éviter tout risque médical. Le choix d’un graphique en barres avec un code couleur est idéal, car il facilite une comparaison facile.

```python
import pandas as pd
import matplotlib.pyplot as plt

df_sauce = pd.read_csv("sauce.csv", sep=";")
valeurs = df_sauce['NOM'].value_counts().head(25)

plt.figure(figsize=(10, 10))
plt.pie(valeurs, labels=valeurs.index, autopct='%1.1f%%')
plt.title("Pourcentages des sauces")
plt.show()
```
