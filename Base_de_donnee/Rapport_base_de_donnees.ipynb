```python
import pandas as pd
import matplotlib.pyplot as plt

df_tenrac = pd.read_csv("tenrac.csv", sep=";")
valeurs = df_tenrac['ID_GENRE'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(valeurs, labels=valeurs.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title("Répartition Homme / Femme / Autre")
plt.show()
```
